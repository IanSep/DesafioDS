### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

def sumaPrecios(Tienda):
  suma_precios=0
  for i in Tienda['Precio']:
    suma_precios +=i
  print(suma_precios)

sumaPrecios(tienda)
sumaPrecios(tienda2)
sumaPrecios(tienda3)
sumaPrecios(tienda4)
```

```python
import matplotlib.pyplot as plt
def sumaPrecios(Tienda):
  suma_precios=0
  for i in Tienda['Precio']:
    suma_precios +=i
  return suma_precios

tienda_names = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
suma_precios_results = [sumaPrecios(tienda), sumaPrecios(tienda2), sumaPrecios(tienda3), sumaPrecios(tienda4)]

plt.figure(figsize=(8, 6))
plt.bar(tienda_names, suma_precios_results, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.xlabel('Tienda')
plt.ylabel('Suma de Precios')
plt.title('Suma de Precios por Tienda')
plt.show()
```

# 2. Ventas por categoría

```python
from collections import Counter
def Conteo(Tienda):
  conteoTienda=Counter(Tienda['Categoría del Producto'])
  print(conteoTienda)
  masComun=conteoTienda.most_common(1)[0]
  print(f"La categoría más común es '{masComun[0]}' con '{masComun[1]} productos.")
print("--Tienda 1--")
Conteo(tienda)
print("******\n")
print("--Tienda 2--")
Conteo(tienda2)
print("******\n")
print("--Tienda 3--")
Conteo(tienda3)
print("******\n")
print("--Tienda 4--")
Conteo(tienda4)
print("******\n")


```

```python
import matplotlib.pyplot as plt
def Conteo(Tienda, title):
  conteoTienda = Counter(Tienda['Categoría del Producto'])
  print(conteoTienda)
  masComun = conteoTienda.most_common(1)[0]
  print(f"La categoría más común es '{masComun[0]}' con '{masComun[1]} productos.")

  labels = list(conteoTienda.keys())
  sizes = list(conteoTienda.values())

  plt.figure(figsize=(8, 8))
  plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
  plt.title(title)
  plt.axis('equal')
  plt.show()

print("--Tienda 1--")
Conteo(tienda, "Distribución de Categorías - Tienda 1")
print("******\n")
print("--Tienda 2--")
Conteo(tienda2, "Distribución de Categorías - Tienda 2")
print("******\n")
print("--Tienda 3--")
Conteo(tienda3, "Distribución de Categorías - Tienda 3")
print("******\n")
print("--Tienda 4--")
Conteo(tienda4, "Distribución de Categorías - Tienda 4")
print("******\n")
```

# 3. Calificación promedio de la tienda


```python
def Calificacion(Tienda):
  for i in range(len(Tienda['Calificación'])):
    promedio=sum(Tienda['Calificación'])/len(Tienda['Calificación'])
  print(promedio)
print("--Tienda 1--")
Calificacion(tienda)
print("--Tienda 2--")
Calificacion(tienda2)
print("--Tienda 3--")
Calificacion(tienda3)
print("--Tienda 4--")
Calificacion(tienda4)
```

```python
import matplotlib.pyplot as plt
def Calificacion(Tienda):
  return sum(Tienda['Calificación'])/len(Tienda['Calificación'])

tienda_names = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificacion_results = [Calificacion(tienda), Calificacion(tienda2), Calificacion(tienda3), Calificacion(tienda4)]

plt.figure(figsize=(8, 6))
plt.bar(tienda_names, calificacion_results, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.title('Calificación Promedio por Tienda')
plt.ylim(0, 5)
plt.show()
```

# 4. Productos más y menos vendidos

```python
def ConteoProducto(Tienda):
  conteoTienda=Counter(Tienda['Producto'])
  print(conteoTienda)
  masComun=conteoTienda.most_common(1)[0]
  menosComun=conteoTienda.most_common()[-1]
  print(f"La categoría más común es '{masComun[0]}' con '{masComun[1]} productos.")
  print(f"La categoría menos común es '{menosComun[0]}' con '{menosComun[1]} productos.")
print("--Tienda 1--")
ConteoProducto(tienda)
print("******\n")
print("--Tienda 2--")
ConteoProducto(tienda2)
print("******\n")
print("--Tienda 3--")
ConteoProducto(tienda3)
print("******\n")
print("--Tienda 4--")
ConteoProducto(tienda4)
print("******\n")
```

```python
import matplotlib.pyplot as plt
def ConteoProducto(Tienda, title):
  conteoTienda = Counter(Tienda['Producto'])
  print(conteoTienda)
  masComun = conteoTienda.most_common(1)[0]
  menosComun = conteoTienda.most_common()[-1]
  print(f"El producto más común es '{masComun[0]}' con '{masComun[1]} unidades vendidas.")
  print(f"El producto menos común es '{menosComun[0]}' con '{menosComun[1]} unidades vendidas.")
  products = list(conteoTienda.keys())
  counts = list(conteoTienda.values())
  plt.figure(figsize=(12, 6))
  plt.bar(products, counts, color='skyblue')
  plt.xlabel('Producto')
  plt.ylabel('Cantidad Vendida')
  plt.title(title)
  plt.xticks(rotation=90)
  plt.tight_layout()
  plt.show()

print("--Tienda 1--")
ConteoProducto(tienda, "Ventas por Producto - Tienda 1")
print("******\n")
print("--Tienda 2--")
ConteoProducto(tienda2, "Ventas por Producto - Tienda 2")
print("******\n")
print("--Tienda 3--")
ConteoProducto(tienda3, "Ventas por Producto - Tienda 3")
print("******\n")
print("--Tienda 4--")
ConteoProducto(tienda4, "Ventas por Producto - Tienda 4")
print("******\n")
```

# 5. Envío promedio por tienda

```python
def CostoEnvio(Tienda):
  for i in range(len(Tienda['Costo de envío'])):
    promedio=sum(Tienda['Costo de envío'])/len(Tienda['Costo de envío'])
  print(promedio)
print("--Tienda 1--")
CostoEnvio(tienda)
print("--Tienda 2--")
CostoEnvio(tienda2)
print("--Tienda 3--")
CostoEnvio(tienda3)
print("--Tienda 4--")
CostoEnvio(tienda4)
```

```python
import matplotlib.pyplot as plt
def CostoEnvio(Tienda):
  return sum(Tienda['Costo de envío'])/len(Tienda['Costo de envío'])

tienda_names = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costo_envio_results = [CostoEnvio(tienda), CostoEnvio(tienda2), CostoEnvio(tienda3), CostoEnvio(tienda4)]

plt.figure(figsize=(8, 6))
plt.bar(tienda_names, costo_envio_results, color=['skyblue', 'lightgreen', 'salmon', 'gold'])
plt.xlabel('Tienda')
plt.ylabel('Costo Promedio de Envío')
plt.title('Costo Promedio de Envío por Tienda')
plt.show()
```
